```python
import matplotlib.pyplot as plt
import pickle

def show_history(history_path: str):
    with open(history_path, "rb") as file:
        data = pickle.load(file)
    fig, (ax1, ax2)= plt.subplots(1, 2)
    fig.set_size_inches(10, 5)

    ax1.set_title("Loss")
    ax1.plot(data["Train loss"], label="train")
    ax1.plot(data["Test loss"], label="test")
    ax1.grid()
    ax1.legend()

    ax2.set_title("Accuracy")
    ax2.plot(data["Train acc"], label="train")
    ax2.plot(data["Test acc"], label="test")
    ax2.grid()
    ax2.legend()
```

```python
show_history("history/model_2025-11-28_00-03-04.pkl")
```
